```python
# Enable inline plotting
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np


# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, label="Sine Wave", color="blue")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Line Chart Example")
ax.legend()

# Show the figure
fig.show()  # Works in standalone scripts, NOT in Colab/Jupyter
```
