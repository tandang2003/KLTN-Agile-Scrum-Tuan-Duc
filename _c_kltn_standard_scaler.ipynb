```python
def calculate_issue_statistics(df):
    categorical_columns = ["type","priority","gunning_fog"]
    df = df.copy(deep = True)
    for col in categorical_columns:
      df[col] = df[col].astype(str).map(eval(f"map_{col}"))
    # Define numerical aggregation functions
    numerical_columns = [
        "no_comment", "no_affectversion", "no_fixversion", "no_issuelink",
        "no_blocking", "no_blockedby", "no_fixversion_change",
        "no_priority_change", "no_des_change"
    ]

    numerical_aggregations = {
        col: ["min", "max", "mean", "median", "std", "var", lambda x: x.max() - x.min()]
        for col in numerical_columns
    }

    # Group by boardId and sprintId
    grouped = df.groupby(["boardid", "sprintid"])

    # Numerical statistics
    numerical_stats = grouped.agg(numerical_aggregations)
    numerical_stats.columns = [
        f"{col}_{agg}" if agg != "<lambda_0>" else f"{col}_range"
        for col, agg in numerical_stats.columns
    ]

    # Replace NaN with 0 in numerical statistics
    numerical_stats = numerical_stats.fillna(0)

    # Categorical statistics: Type, Priority, and Gunning Fog
    categorical_columns = ["type", "priority", "gunning_fog"]
    categorical_stats = []

    for col in categorical_columns:
        # Include NaN values in the frequency count
        categorical_frequency = (
            grouped[col]
            .value_counts(dropna=False)  # Include NaN values
            .unstack(fill_value=0)
        )
        categorical_frequency.columns = [f"{col}_{val}_frequency" for val in categorical_frequency.columns]
        categorical_stats.append(categorical_frequency)

    # Combine numerical and categorical statistics
    result = numerical_stats.reset_index()
    for cat_stat in categorical_stats:
        result = result.merge(cat_stat.reset_index(), on=["boardid", "sprintid"], how="left")

    # Replace NaN with 0 in categorical statistics as well
    result = result.fillna(0)

    return result
```

## **BOW method**

```python
def bow (df, k):
  df = df.copy(deep = True)
  for col in categorical_columns:
    df[col] = df[col].astype(str).map(eval(f"map_{col}"))
  X = df.to_numpy()
  kmeans = KMeans(n_clusters = k, random_state=42, n_init=10)
  kmeans.fit(X)
  centers = kmeans.cluster_centers_

  df['Cluster'] = kmeans.predict(X)
  df = df[["boardid", "sprintid","Cluster"]]
  grouped = df.groupby(["boardid", "sprintid", "Cluster"]).size().reset_index(name="count")


  # Pivot the table
  df_pivot = grouped.pivot_table(index=["boardid", "sprintid"], columns="Cluster", values="count", fill_value=0)

  # Reset index to make it a regular DataFrame
  df_pivot.reset_index(inplace=True)

  # Rename columns
  df_pivot.columns.name = None  # Remove multi-index column name

  df_pivot = df_pivot.astype('Int64')
  return df_pivot
```

## **Prepare data**

```python
def merge(df1, df2):
    df = pd.merge(df1, df2, on=["boardid", "sprintid"], how="left")
    df.fillna(0, inplace=True)
    df.columns = df.columns.astype(str)  # Fixed typo here
    return df
```

```python
def seperate_X_y(df):
  X = df.drop(columns=['vel_diff',"boardid", "sprintid", "name"])
  y = df['vel_diff']
  return X,y
```

```python
def get_data_statistic_for_regression(project):
  time_predict=[0,30,50,80]
  dataframes = {}
  for predict_time in time_predict:
    data_iteration = data.get_data("iterator", predict_time, project)
    data_issues = data.get_data("issues", predict_time, project)
    data_statistics = calculate_issue_statistics(data_issues)
    dataframes[predict_time] = merge(data_iteration, data_statistics)
  return dataframes

def get_data_bow_for_regression(project):
  time_predict=[0,30,50,80]
  dataframes = {}
  for predict_time in time_predict:
    data_iteration = data.get_data("iterator", predict_time, project)
    data_issues = data.get_data("issues", predict_time, project)
    data_bow = bow(data_issues, 100)
    dataframes[predict_time] = merge(data_iteration, data_bow)
  return dataframes

def get_data_statistics_bow_for_regression(project):
    time_predict=[0,30,50,80]
    dataframes = {}
    for predict_time in time_predict:
      data_iteration = data.get_data("iterator", predict_time, project)
      data_issues = data.get_data("issues", predict_time, project)
      data_bow = bow(data_issues, 100)
      data_statistics = calculate_issue_statistics(data_issues)
      dataframes[predict_time] = merge(data_iteration, merge(data_statistics, data_bow))
    return dataframes
```

```python
def label_vel_diff(df):
  df_iterator = df.copy(deep=True)
  label = "label_vel_diff"
  under =  np.percentile(df_iterator['vel_diff'], 33)
  over = np.percentile(df_iterator['vel_diff'], 66)
  for i, row in df_iterator.iterrows():
    value = row["vel_diff"]
    if value < under:
      df_iterator.loc[i, label] = -1
    elif value > over:
      df_iterator.loc[i, label] = 1
    else:
      df_iterator.loc[i, label] = 0
  return df_iterator
```

```python
def plot_confusion_matrix_with_metrics(y_true, y_pred, labels, title="Confusion Matrix"):
    """
    Plot a beautiful confusion matrix using Seaborn and display classification metrics in a table.

    Parameters:
    - y_true: True labels
    - y_pred: Predicted labels
    - labels: Class names for better readability
    - title: Title for the plot
    """

    # Compute Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)

    # Compute Classification Report
    report = classification_report(y_true, y_pred, target_names=labels, output_dict=True)

    # Compute Accuracy
    accuracy = accuracy_score(y_true, y_pred)

    # Extract Metrics for Each Class
    table_data = []
    for label in labels:
        metrics = report[label]
        table_data.append([label, f"{metrics['precision']:.2f}", f"{metrics['recall']:.2f}", f"{metrics['f1-score']:.2f}", int(report[label]["support"])])

    # Add Macro & Weighted Averages
    macro_avg = report["macro avg"]
    weighted_avg = report["weighted avg"]
    table_data.append(["Macro Avg", f"{macro_avg['precision']:.2f}", f"{macro_avg['recall']:.2f}", f"{macro_avg['f1-score']:.2f}", "-"])
    table_data.append(["Weighted Avg", f"{weighted_avg['precision']:.2f}", f"{weighted_avg['recall']:.2f}", f"{weighted_avg['f1-score']:.2f}", "-"])

    # Print Metrics Table
    print("\nClassification Metrics:\n")
    print(tabulate(table_data, headers=["Class", "Precision", "Recall", "F1-Score", "Support"], tablefmt="fancy_grid"))
    print(f"\nOverall Accuracy: {accuracy:.4f}\n")

    # Create Confusion Matrix Plot
    plt.figure(figsize=(7, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels,
                yticklabels=labels, linewidths=1.5, linecolor="black",
                cbar=True, annot_kws={"size": 14})

    # Labels & Title
    plt.xlabel("Predicted Label", fontsize=12)
    plt.ylabel("True Label", fontsize=12)
    plt.title(title, fontsize=14)
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)

    # Show Plot
    plt.show()
    return accuracy, metrics
```
